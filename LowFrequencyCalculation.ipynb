Low Frequency Calculation File

Takes in the processed data and creates a file with frequency of words from all questions

Imports

```python
!pip install pandas

from pandas.api.types import is_string_dtype
from pandas.api.types import is_numeric_dtype

import numpy as np
import pandas as pd
import nltk
from nltk.corpus import stopwords
import re

nltk.download('punkt')
nltk.download('stopwords')
```

Read data into dataframe

```python
inputFile = 'Language_Processed_Data.csv'
inputData = pd.read_csv(inputFile, sep=',', skiprows=(0), header=(0))

inputData['Question'].dropna(inplace=True)
inputData.reset_index(inplace=True)
```

Extracting Sentences

```python
#Extracting questions from data input file
sentences = []

for index, row in inputData.iterrows():
  sentences.append(row['Question'])
```

Processing data

```python
#Tokenizing sentences into words
corpus = []

for s in sentences:
  tokens = nltk.word_tokenize(s)
  for t in tokens:
    t = ''.join([i for i in t if not i.isdigit()])
    t = t.strip()

    if ' ' in t:
      words = t.split("\\s")
      for w in words:
        if not w.isdigit():
          corpus.append(str(w))
    else:
      if not t.isdigit():
        corpus.append(str(t))



#Remove punctuation and extra white spaces
for i in range(len(corpus)):
  corpus[i] = corpus[i].lower()
  corpus[i] = re.sub(r'\W', ' ', corpus[i])
  corpus[i] = re.sub(r'\s+', ' ', corpus[i])
 
```

Count number of words

```python
wordFrequency = {}

#Counting the frequency of words
for word in corpus:
  if not word.isdigit():
    wordFrequency[word] = wordFrequency.get(word, 0) + 1

#Sorting words from lower to higher frequency
wordFrequency = {k: v for k, v in sorted(wordFrequency.items(), key=lambda item: item[1])}

#Writing word frequency into output file
with open('wordFrequency.txt', 'w') as outfile:
  for key in wordFrequency:
    outfile.write(key + ': ' + str(wordFrequency[key]) + '\n')
```
